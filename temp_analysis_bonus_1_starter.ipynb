### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
june=df[df.index.month==6]
june.head(30)
```

```python
december= df[df.index.month==12]
december.head(30)
```

```python
# Identify the average temperature for June
juneavg= june['tobs'].mean()
print(juneavg)
```

```python
# Identify the average temperature for December
decavg=december['tobs'].mean()
print(decavg)
```

```python
# Create collections of temperature data
junet= june.tobs
junet
```

```python
dect=december.tobs
dect
```

```python
# Run paired t-test
stats.ttest_ind(dect,junet)
```

### Analysis

A 3.9 degrees difference in average temperature in June than December indicates a slight but very little variance based on the time of year visiting Hawaii.
